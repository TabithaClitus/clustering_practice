```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
X,_=make_blobs(n_samples=300, centers=3, cluster_std=0.7, random_state=42)  
```

```python
kmeans=KMeans(n_clusters=3, random_state=42)
#Fit: runs the KMeans algorithm on the dataset X and Predict: assigns each point to a cluster (0,1,2)
kmeans_labels=kmeans.fit_predict(X)   
plt.scatter(X[:,0],X[:,1],c=kmeans_labels,cmap='viridis',s=50)  #c --> colors the points by cluster, s --> marker size
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='red',marker='X',s=200)  #this is for the cross symbol in center
plt.title("KMeans Clustering")
plt.show()
```

```python
Z=linkage(X,method='ward')  #creates the hierarchical tree (dendrogram structure)
#fcluster (flat cluster function) → it cuts the dendrogram into actual cluster labels, t--> no.of clusters
hier_labels=fcluster(Z,t=3,criterion='maxclust')
plt.scatter(X[:,0],X[:,1],c=hier_labels,cmap='prism',s=50)
plt.title("Hierarchical Clustering")
plt.show()

dendrogram(Z)
plt.title("Dendrogram")
plt.show()
```
